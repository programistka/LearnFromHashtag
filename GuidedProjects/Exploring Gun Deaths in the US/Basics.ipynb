```python
import csv
```

```python
f = open("guns.csv")
```

```python
data = list(csv.reader(f))
```

```python
print(data[0:5])
```

```python
headers = data[0]
data = data[1:]
```

```python
years = []
for row in data:
    years.append(row[1])
```

```python
print(years)
```

```python
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] += 1;
    else:
        year_counts[year] = 1;
```

```python
print(year_counts)
```

```python
import datetime;
dates = []
for d in data:
    dates.append(datetime.datetime(year=int(d[1]), month=int(d[2]), day = 1))
```

```python
print(dates[0:5])
```

```python
date_counts = {}
for d in dates:
    if d in date_counts:
        date_counts[d]=+1;
    else:
        date_counts[d]=1;
```

```python
print(date_counts);
```

```python
intents = [];
for item in data:
    intents.append(item[3]);
```

```python
races = [];
for item in data:
    races.append(item[7]);
```

```python
homicide_race_per_hundredk = {}
for i, race in enumerate(races):
    if race not in homicide_race_per_hundredk:
        homicide_race_per_hundredk[race] = 0
    if intents[i] == "Homicide":
        homicide_race_per_hundredk[race] += 1
```

```python
print(homicide_race_per_hundredk);
```
