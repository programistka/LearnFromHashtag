```python
import csv
```

```python
f = open("guns.csv")
```

```python
data = list(csv.reader(f))
```

```python
print(data[0:5])
```

```python
headers = data[0]
data = data[1:]
```

```python
years = []
for row in data:
    years.append(row[1])
```

```python
print(years)
```

```python
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] += 1;
    else:
        year_counts[year] = 1;
```

```python
print(year_counts)
```

```python
import datetime;
dates = []
for d in data:
    dates.append(datetime.datetime(year=int(d[1]), month=int(d[2]), day = 1))
```

```python
print(dates[0:5])
```

```python
date_counts = {}
for d in dates:
    if d in date_counts:
        date_counts[d]=+1;
    else:
        date_counts[d]=1;
```

```python
print(date_counts);
```

```python
sex_counts = {}
for item in data:
    if item[5] in sex_counts:
        sex_counts[item[5]] += 1;
    else:
        sex_counts[item[5]] = 1;
```

```python
print(sex_counts);
```

```python
race_counts = {}
for item in data:
    if item[7] in race_counts:
        race_counts[item[7]] += 1;
    else:
        race_counts[item[7]] = 1;
```

```python
print(race_counts);
```

```python
intents = [];
for item in data:
    intents.append(item[3]);
```

```python
races = [];
for item in data:
    races.append(item[7]);
```

```python
homicide_race_per_hundredk = {}
for i, race in enumerate(races):
    val = intents[i]
    if(val == 'Homicide'):
        if(race in homicide_race_per_hundredk):
            homicide_race_per_hundredk[race] += 1
        else:
            homicide_race_per_hundredk[race] = 1
```
