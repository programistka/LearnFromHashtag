```python
def read_csv(filename):
    f = open(filename).read()
    splitted = f.split("\n")
    string_list = splitted[1:len(splitted)]
    final_list = []
    for element in string_list:
        int_fields = []
        string_fields = element.split(',')
        for field in string_fields:
            print(field)
            int_fields.append(int(field))
        final_list.append(int_fields)
    return final_list
```

```python
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
```

```python
print(cdc_list[0:10])
```

```python
def month_births(lists):
    births_per_month = {}
    for list in lists:
        month = int(list[1])
        births = int(list[4])
        if month in births_per_month:
            births_per_month[month] += births
        else:
            births_per_month[month] = births
    return births_per_month

    
```

```python
cdc_month_births = month_births(cdc_list)
```

```python
print(cdc_month_births)
```

```python
def dow_births(lists):
    births_per_day_of_the_week = {}
    for list in lists:
        day_of_week = int(list[3])
        births = int(list[4])
        if day_of_week in births_per_day_of_the_week:
            births_per_day_of_the_week[day_of_week] += births
        else:
            births_per_day_of_the_week[day_of_week] = births
    return births_per_day_of_the_week
```

```python
cdc_day_births = dow_births(cdc_list)
```

```python
print(cdc_day_births)
```

```python
def calc_counts(data, column):
    totals = {}
    for list in data:
        col = int(list[column])
        births = int(list[4])
        if col in totals:
            totals[col] += births
        else:
            totals[col] = births
    return totals
```

```python
cdc_year_births = calc_counts(cdc_list, 0)
```

```python
print(cdc_year_births)
```

```python
cdc_month_births = calc_counts(cdc_list, 1)
```

```python
print(cdc_month_births)
```
